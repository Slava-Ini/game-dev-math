# Sine, Cosine and Tangent

## Tangent

_Tangent_ is a ratio of the side opposite to $\theta$ to the side adjacent to $\theta$

The formula for the triangle with legs 3 (opposite) and 4 (adjacent) is $$tan \theta = \frac{3}{4} = 0.75 = 36.87\degree$$

- $tan(45\degree) = 1$
- $tan(0\degree) = 0$
- $tan(\theta) = \frac{sin\theta}{cos\theta}$
- $tan(90\degree)$ is impossible because $sin(90\degree)=1$ and $cos(90\degree) = 0$ - we can't divide by $0$
- _Tangent_ has no finite max value, because when approaching to $90\degree$ and $270\degree$ it approaches _asymptote_ i.e. never touches the values, in other words it's undefined in a range ($\frac{\pi}{2}$ to $\frac{3\pi}{2}$, or $90\degree$ to $270\degree$)

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np
import math

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a

def cosine(x):
    a = []
    for i in x:
        a.append(math.tan(i))
    return a

# Generate numbers (fromt, to, step_    
x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)

y = cosine(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottom x-axis to centre, passing throughv (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,y)
plt.show()
```
