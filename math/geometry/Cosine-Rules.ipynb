# Cosine Rules

Cosine rules (or laws or cosine) is a several formulas that can be used to determine the length of a side of a triangle to the opposite angle cosine

Following is the simple reasoning behind it (using Pythagoras):
   

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Triangle
vertices = [(2, 2), (8, 2), (5, 9)]
triangle = Polygon(vertices, closed=True, facecolor="none", edgecolor="black", alpha=0.5)

s_vertices = [(5, 2.2), (5.2, 2.2), (5.2,  2)]
square = Polygon(s_vertices, closed=False, facecolor="none", edgecolor="black", alpha=0.5)

vertex_labels = ["A", "B", "C"]
for i, (x, y) in enumerate(vertices):
    plt.text(x, y, vertex_labels[i], fontsize=12, ha="right", va="bottom")

plt.gca().add_patch(triangle)
plt.gca().add_patch(square)

plt.annotate(r'$b$', xy=(3, 5), xytext=(0, 0), textcoords='offset points')
plt.annotate(r'$a$', xy=(7, 5), xytext=(0, 0), textcoords='offset points')
plt.annotate(r'$c$', xy=(5.2, 1.5), xytext=(0, 0), textcoords='offset points')
plt.annotate("D", xy=(4.8, 1.5), xytext=(0, 0), textcoords='offset points')
plt.annotate(r'$x$', xy=(3.5, 2.5), xytext=(0, 0), textcoords='offset points')
plt.annotate(r'$c-x$', xy=(6.2, 2.5), xytext=(0, 0), textcoords='offset points')
plt.annotate(r'$h$', xy=(4.6, 5), xytext=(0, 0), textcoords='offset points')

plt.plot([5, 5], [2, 9], c="gray")


plt.xlim(1, 10)
plt.ylim(1, 10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

1. $\triangle{BCD}$ :  
   $a^2 = h^2 + (c-x)^2$  
   $a^2 = h^2 + c^2 - 2cx + x^2$
   
2. $\triangle{ACD}$ :  
   $b^2 = x^2 + h^2$  
   $\color{lightgreen}{h^2 = b^2 - x^2}$

   
3. Following the first formula  :  
   $a^2 = \color{lightgreen}(b^2 - x^2) \color{white}+ c^2 - 2cx + x^2$  
   $\color{purple}a^2 = b^2 + c^2 - 2cx$

   
4. $\triangle{ACD}$ :  
   $cos(A) = \dfrac{x}{b}$  
   $cos\theta = \dfrac{adjacent\space angle}{hypotenuse}$  
   $\color{yellow}x = b\times cos(A)$

5. Final formula:  
   $a^2 = \color{purple}b^2 + c^2 - 2c \times \color{yellow}b \times cos(A)$

   All formulas:

   $$a^2 = b^2 + c^2 - 2bc \times cos(A)$$
   $$c^2 = a^2 + b^2 - 2ab \times cos(C)$$
   $$b^2 = a^2 + c^2 - 2ac \times cos(B)$$
